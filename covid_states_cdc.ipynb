```python
import pandas as pd     #(version 1.0.0)
import plotly           #(version 4.5.4) #pip install plotly==4.5.4
import plotly.express as px
import plotly.io as pio
import seaborn as sns
```

```python
data = pd.read_csv('./United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv')
data['date'] = pd.to_datetime(data['submission_date'])
data.sort_values("date", inplace=True)
data = data.convert_dtypes()
data.info()
```

```python
data.isna().sum()
```

```python
ca = data[data["state"] == "CA"]
nyc = data[data["state"] == "NYC"]
fl = data[data["state"] == "FL"]
tx = data[data["state"] == "TX"]
la = data[data["state"] == "LA"]
big_states_df = data[data["state"].isin(["CA", "NYC", "FL", "TX", "LA"])]
```

```python
px.line(big_states_df, x="date", y="tot_cases", title="Covid-19 Total Cases", color="state")
```

```python
px.line(big_states_df, x="date", y="tot_death", title="Covid-19 Death Count", color="state")
```
