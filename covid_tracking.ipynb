```python
import pandas as pd     #(version 1.0.0)
import plotly           #(version 4.5.4) #pip install plotly==4.5.4
import plotly.express as px
import plotly.io as pio
```

```python
data = pd.read_csv('./covid_daily_all-states-history.csv')
```

```python
ca = data[data["state"] == "CA"]
big_states_df = data[data["state"].isin(["CA", "NY", "FL", "TX", "LA"])]
```

```python
px.line(big_states_df, x="date", y="positive", title="Covid-19 Positive Count", color="state")
```
